```python
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import math

n = 100
k = 50
w = 14

cnt = 0

result = [ [0]*40 for i in range(40)]

for l in range(n-k-w+1):
    if w > k+l:
        x = k+l
    else:
        x = w
    for p in range(x+1):
        L = math.comb((int)((k+l)/2), (int)(p/4))
        q = int(np.log2(math.comb(p, (int)(p/2))))
        A = max(L, math.pow(L,2)/math.pow(2, q), math.pow(L,4)/math.pow(2, q+l))
        B = math.comb(n-k-l, w-p)
        C = math.comb(k+l, p)
        complexity = np.log10(A/(B*C))
        #if complexity > 1:
        #    complexity = 0
        #    cnt += 1
        result[l][p] = complexity


#print("too big val:", cnt)

data = np.array(result)
plt.imshow(data, interpolation='none')
plt.colorbar(orientation='vertical')
plt.show()
```
